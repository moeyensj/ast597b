### Part 2:

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy import ndimage

% matplotlib inline
```

```python
data = fits.getdata('data.fit')
```

```python
condition_blue = (data['dered_r'] > 20.0)*(data['dered_r'] <= 20.66)
condition_green = (data['dered_r'] > 20.66)*(data['dered_r'] <= 21.33) 
condition_red = (data['dered_r'] > 21.33)*(data['dered_r'] <= 22.0) 
```

```python
blue = data[condition_blue]
green = data[condition_green]
red = data[condition_red]
```

```python
xbins = (230-100)/0.5
ybins = (60--5)/0.5
aspect_ratio = ybins/xbins

b, bx, by = np.histogram2d(blue['ra'], blue['dec'], bins=[xbins,ybins])
g, gx, gy = np.histogram2d(green['ra'], green['dec'], bins=[xbins,ybins])
r, rx, ry = np.histogram2d(red['ra'], red['dec'], bins=[xbins,ybins])

print b.max()
print g.max()
print r.max()
```

### Part 3:

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10,10*aspect_ratio)

ax.set_title('Field of Streams')
ax.set_ylabel(r'DEC ($\delta$)')
ax.set_xlabel(r'RA ($\alpha$)')

rgb = np.zeros((xbins,ybins,3),dtype='uint8')
rgb[:,:,0] = r
rgb[:,:,1] = g
rgb[:,:,2] = b

rgb_rotated = ndimage.rotate(rgb, 270)
ax.imshow(rgb_rotated,cmap=plt.cm.brg,origin='lower',extent=[230,100,-5,60]);
```
