```python
# Query used to acquire data
#      SELECT ra, dec, dered_g, dered_r into mydb.MyTable from Star
#      WHERE (ra > 225) AND (ra < 235) AND (dec > -2.5) AND (dec < 4.0)

hdulist = fits.open('HW1_P2.fit')
dataFits = hdulist[1].data
dataKeys = ['ra','dec','dered_g','dered_r']

try:
    os.remove('HW1_P2.db')
except OSError:
    pass

con = sqlite3.connect('HW1_P2.db')
data = pd.DataFrame(dataFits, columns=dataKeys)
data.to_sql('HW1_P2',con)
```

```python
fig, ax = plt.subplots(1,1)

ax.scatter(dataFits['ra'],dataFits['dec'],s=0.0005)

ax.set_xlim(225,235);
ax.set_ylim(-2.5,4);
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$\delta$')
ax.set_title('RA vs DEC')
ax.invert_xaxis();

# Region 1
r1 = [228.6,229.4,-0.4,0.4]
ax.hlines([r1[2],r1[3]],r1[0],r1[1],color='w');
ax.vlines([r1[0],r1[1]],r1[2],r1[3],color='w');

# Region 2
r2 = [229.6,230.4,-0.4,0.4]
ax.hlines([r2[2],r2[3]],r2[0],r2[1],color='w');
ax.vlines([r2[0],r2[1]],r2[2],r2[3],color='w');
```

Select a subset of the data in a region (a rectangle is fine!) tightly bounding the observed object. Select data in another, equally sized region, nearby but *away* from the object. Do the selection in two ways: with an SQL query of your local sqlite database, and using the indexing operators in Python (the square brackets '[]'). Show that the two give identical results.

```python
# Region 1: ra: 228.6,229.4 dec: -0.4,0.4
# SQL
dataR1_SQL = pd.read_sql("""
    SELECT 
        HW1_P2.ra, HW1_P2.dec, HW1_P2.dered_g, HW1_P2.dered_r
    FROM 
        HW1_P2
    WHERE
        (HW1_P2.ra > 228.6) AND (HW1_P2.ra < 229.4) AND (HW1_P2.dec > -0.4) 
        AND (HW1_P2.dec < 0.4) AND (HW1_P2.dered_g != -9999) AND (HW1_P2.dered_r != -9999)
""", con)

condR1 = ((dataFits['ra'] > r1[0])*(dataFits['ra'] < r1[1])*(dataFits['dec'] > r1[2])*
          (dataFits['dec'] < r1[3])*(dataFits['dered_g'] != -9999)*(dataFits['dered_r'] != -9999))
dataR1_PY = dataFits[condR1]

print 'Region 1:'
for i,j in enumerate(dataKeys):
    array1 = np.array(dataR1_SQL[j])
    array2 = np.array(dataR1_PY[j])
    print 'SQL_' + j + ' == ' + 'PY_' + j + '?'
    print all(array1 == array2)
    
# Region 2: ra: 229.6,230.4 dec: -0.4,0.4
dataR2_SQL = pd.read_sql("""
    SELECT 
        HW1_P2.ra, HW1_P2.dec, HW1_P2.dered_g, HW1_P2.dered_r
    FROM 
        HW1_P2
    WHERE
        (HW1_P2.ra > 229.6) AND (HW1_P2.ra < 230.4) AND (HW1_P2.dec > -0.4) AND (HW1_P2.dec < 0.4)
        AND (HW1_P2.dec < 0.4) AND (HW1_P2.dered_g != -9999) AND (HW1_P2.dered_r != -9999)
""", con)

condR2 = ((dataFits['ra'] > r2[0])*(dataFits['ra'] < r2[1])*(dataFits['dec'] > r2[2])*
          (dataFits['dec'] < r2[3])*(dataFits['dered_g'] != -9999)*(dataFits['dered_r'] != -9999))
dataR2_PY = dataFits[condR2]

print
print 'Region 2:'
for i,j in enumerate(dataKeys):
    array1 = np.array(dataR2_SQL[j])
    array2 = np.array(dataR2_PY[j])
    print 'SQL_' + j + ' == ' + 'PY_' + j + '?'
    print np.all(array1 == array2)
```
